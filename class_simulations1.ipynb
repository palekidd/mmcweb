```python
import random

# Bag of marbles
marbles = ['blue'] * 5 + ['red'] * 2 + ['yellow'] * 3 + ['green'] * 1

# Function to pick two random marbles
def pick_marbles():
    return random.sample(marbles, 2)

# Simulate picking two marbles
selected_marbles = pick_marbles()
print(f"The two randomly selected marbles are: {selected_marbles[0]} and {selected_marbles[1]}")
```

```python
import random

# Define the marbles in the bag
bag_of_marbles = ['blue'] * 4 + ['red'] * 17 + ['yellow'] * 1 + ['purple'] * 7 + ['black'] * 5

# Randomly pull 2 marbles without replacement
pulled_marbles = random.sample(bag_of_marbles, 2)

# Display the pulled marbles
print("Pulled marbles:", pulled_marbles)
```

```python
import random
import matplotlib.pyplot as plt
from collections import Counter

# Define the marbles in the bag
bag_of_marbles = ['blue'] * 4 + ['red'] * 17 + ['yellow'] * 1 + ['purple'] * 7 + ['black'] * 5

# Run the selection 50 times
results = []
for _ in range(50):
    pulled_marbles = tuple(random.sample(bag_of_marbles, 2))  # Create a tuple of the pulled marbles
    results.append(pulled_marbles)

# Count the frequency of each pair
counter = Counter(results)

# Create labels and counts for the chart
labels, counts = zip(*counter.items())

# Create a bar chart
plt.figure(figsize=(10, 5))
plt.bar([str(label) for label in labels], counts, color='skyblue')
plt.xticks(rotation=90)  # Rotate labels for better readability
plt.xlabel('Marble Pairs')
plt.ylabel('Frequency')
plt.title('Frequencies of Marble Pairs Pulled from the Bag (50 trials)')
plt.tight_layout()
plt.show()
```
