```python
import matplotlib.pyplot as plt
import numpy as np

# Define slope (m) and y-intercept (c)
m = 0.715
c = 50.24

# Create x values
x = np.linspace(0, 10, 50)

# Calculate corresponding y values
y = m * x + c

# Plot the line
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = 2x + 5', color='blue')
plt.axhline(0, color='black', linewidth=0.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis
plt.grid(True)
plt.title("Line: y = mx + c")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Data: Father and Son heights (cm)
father_cm = [
    165.1, 160.78, 165.1, 167.13, 155.19, 160.02, 166.12, 164.34, 167.89, 170.18,
    149.86, 159.77, 161.8, 162.81, 164.34, 165.61, 168.66, 171.2, 169.67, 172.21,
    176.53, 158.75, 162.05, 163.83, 165.1, 164.34, 166.37, 166.62, 169.93, 169.16,
    168.91, 172.47, 173.48, 171.96, 173.99, 176.28, 182.37, 158.5, 163.83, 162.31,
    166.12, 165.1, 166.88, 174.5, 168.91, 164.85, 176.78, 164.59, 165.86, 167.13,
    166.62, 167.39, 167.13, 170.43, 171.2, 170.94, 169.16, 173.23, 173.48, 172.21
]

son_cm = [
    151.89, 160.53, 160.78, 159.51, 163.32, 163.07, 162.81, 162.56, 164.08, 162.56,
    165.61, 166.12, 166.88, 166.12, 165.86, 164.59, 165.1, 165.35, 167.64, 165.35,
    166.37, 169.16, 168.66, 167.89, 167.64, 167.64, 166.88, 167.64, 168.4, 170.69,
    156.72, 167.13, 167.89, 167.89, 168.4, 168.4, 168.66, 170.69, 171.45, 169.67,
    170.18, 169.67, 176.02, 175.51, 156.72, 166.88, 175.26, 171.2, 170.69, 170.94,
    171.7, 171.45, 169.93, 171.7, 169.67, 170.69, 171.45, 171.7, 171.45, 171.45
]

# Linear regression using scipy
slope, intercept, r_value, p_value, std_err = linregress(father_cm, son_cm)

# Predicted line values
son_predicted = [slope * x + intercept for x in father_cm]

# Plotting
plt.figure(figsize=(8,6))
plt.scatter(father_cm, son_cm, color='blue', label='Data')
plt.plot(father_cm, son_predicted, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
plt.title("Father vs Son Heights with Linear Regression")
plt.xlabel("Father Height (cm)")
plt.ylabel("Son Height (cm)")
plt.legend()
plt.grid(True)
plt.show()

# Print r and R²
print(f"Correlation coefficient (r): {r_value:.3f}")
print(f"Coefficient of determination (R²): {r_value**2:.3f}")
```
